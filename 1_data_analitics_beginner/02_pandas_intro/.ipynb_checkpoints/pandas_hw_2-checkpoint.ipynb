## Task 1

### Load `Fishing.csv`. Show first 8 rows.

```python
df = pd.read_csv("Fishing.csv")
df[0:8]
```

## Task 2

### Add `log_income`, using `.apply() method`. Will contain natural log of income.

```python
df['log_income'] = df['income']
df['log_income'] = df['log_income'].apply(np.log)
df.head()
```

## Task 3

### Count for each respondend absolute value of difference of price and pbeach and save result in `pdiff`.

```python
absolute = lambda x, y : abs(x-y) 
```

```python
df['pdiff'] = absolute(df['price'], df['pbeach'])
df.head()
```

## Task 4

### Group observations in the table by type of fishing `mode` and display for each type an average price respondents paid for the fishing.

```python
df.groupby('mode').agg('mean')['price']
```

## Task 5

### Group observations by the type of the fishing `mode` and display for each type difference between mean and median value of price, which respondents paid for the fishing.

```python
difference = lambda x, y: abs(x - y)
```

```python
ndf = df.groupby('mode').agg(['mean', 'median'])['price']
ndf
```

```python
difference(ndf['mean'], ndf['median'])
```

## Task 6 

### Group observations by the type of the fishing `mode` and save each type in different dataframe. Save each dataframe in different `.csv` files.

```python
for name, data in df.groupby("mode"):
    data.to_csv(name + '.csv')
```

## Task 6

### Using `.apply()`, create a column `cc_round` with rounded `cc` values until one digit after a floating point.

```python
df2['cc_round'] = df2['cc']
roundl = lambda x: np.round(x, 1)
df2['cc_round'] = df2['cc_round'].apply(roundl)
df2.head()
```

## Task 7

### Add column `fh_status`, which will have types of countries depending on value of Freedom House index.

```python
def fh_status(value):
    if value <= 2.5:
        return "free"
    if 2.5 < value <= 5.0:
        return "partly free"
    if 5.0 < value <= 7.0:
        return "not free"
```

```python
df2['fh_status'] = df2['fh'].apply(fh_status)
df2
```
