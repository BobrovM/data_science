# CSV

```python
import pandas as pd
```

```python
df = pd.read_csv("world-data-2023.csv")
```

```python
df.head()
```

```python
df['Population'] = df['Population'].replace(',','',regex=True).fillna('0').astype(int)
```

```python
new_df = df[df['Population'] > 50000000]
```

```python
new_df
```

```python
new_df.to_csv('result.csv', index_label='index', sep=';')
```
