```python
loans_by_activity_sector = df_kiva_loans[df_kiva_loans['sector'].isin(sectors[:9].index.values)]\
    .groupby(['sector', 'activity'])['loan_amount'].count().reset_index()
```

```python
fig, axes = plt.subplots(3,3, sharex=False, squeeze=False, figsize=(30, 20))

for ax, sec in zip(axes.ravel(), loans_by_activity_sector.sector.unique()):
    temp_df = loans_by_activity_sector[loans_by_activity_sector.sector.eq(sec)]\
        .dropna()\
        .sort_values(by='loan_amount')[-10:]
    ax.set_title(sec, fontsize=20)
    ax.yaxis.label.set_visible(False)
    plt.tight_layout()
    
    temp_df.plot.barh(x='activity', ax=ax, legend=None, fontsize=16)
```

### Trimmed data by 3 standart deviations

```python
loan_amount_median = df_kiva_loans['loan_amount'].median()
loan_amount_std = df_kiva_loans['loan_amount'].std()
```

```python
df_kiva_loans_trimmed = df_kiva_loans[loan_amount_median+\
        loan_amount_std*3 > df_kiva_loans['loan_amount']]
```

```python
plt.figure(figsize=(20,10))
plt.title("Loan amount distribution", fontsize=16)
plt.tick_params(labelsize=14)

sns.histplot(df_kiva_loans_trimmed['loan_amount'], kde=True)
plt.show()
```

## Loan amount distribution by sectors

```python
plt.figure(figsize=(20,10))
plt.title("Loan amount distribution by sectors", fontsize=16)
plt.tick_params(labelsize=14)
plt.xticks(np.arange(0, max(df_kiva_loans['loan_amount']), 10000))

sns.boxplot(x='loan_amount', y='sector', data=df_kiva_loans).set(ylabel=None, xlabel=None)
plt.show()
```
